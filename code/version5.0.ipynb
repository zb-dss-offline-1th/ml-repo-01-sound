```python
import numpy as np
import pandas as pd
import scipy, scipy.spatial
import matplotlib.pyplot as plt
import IPython.display as ipd
import librosa, librosa.display
import speechpy
from pydub import AudioSegment
from matplotlib.patches import ConnectionPatch
from sklearn.preprocessing import minmax_scale
from statistics import mean
%matplotlib inline
```

```python
def sound_diff(path, ref, target, sr, word, phoneme):
    
    def getTrim(y_comp):
        y_comp_pre = librosa.effects.preemphasis(y_comp)
        o_env = librosa.onset.onset_strength(y=y_comp, sr=sr,
                                                    aggregate=np.median,
                                                    fmax=8000, n_mels=256)
        times = librosa.times_like(o_env, sr=sr)
        y_comp_tmp, idx = librosa.effects.trim(y=y_comp_pre, top_db=55, ref=1)
        trim_samples = []
        trim_samples.append(max(0, idx[0]-512))
        trim_samples.append(min(y_comp.shape[0], idx[-1]+512))

        D = np.abs(librosa.stft(y_comp))
        fig, ax = plt.subplots(nrows=3, sharex=True, figsize=(12,8))
        librosa.display.specshow(librosa.amplitude_to_db(D, ref=np.max),
                                x_axis='time', y_axis='log', ax=ax[0])
        ax[0].set(title='Power spectrogram')
        ax[0].label_outer()
        ax[1].plot(times, o_env)
        ax[1].vlines(times[idx//512], 0, o_env.max(), color='r', alpha=0.9,
                linestyle='--', label='Trim')
        ax[1].legend()
        librosa.display.waveshow(y_comp, x_axis="time", ax=ax[2])
        ax[2].set(title='Wave Plot')
        ax[2].label_outer()
        ax[2].vlines(times[idx//512], min(y_comp), y_comp.max(), color='r', alpha=0.9,
                linestyle='--', label='Trim')
        
        return y_comp[trim_samples[0]:trim_samples[-1]]

    def getMFCC(signal, n_mfcc, sr, hop_length, n_fft):
        mfccs = librosa.feature.mfcc(y=signal, sr=sr, hop_length=hop_length, n_mfcc=n_mfcc, n_fft=n_fft)
        mfccs = mfccs[1:,]
        delta_mfccs = librosa.feature.delta(mfccs)
        delta2_mfccs = librosa.feature.delta(mfccs, order=2)
        comprehensive_mfccs = np.concatenate((mfccs, delta_mfccs, delta2_mfccs))
        return comprehensive_mfccs
        
    def amplitude_envelope(signal, n_fft, hop_length):
        return np.array([max(signal[i:i+n_fft]) for i in range(0, len(signal), hop_length)])
    
    def pad(ae, target_len):
        return np.pad(ae, (0, target_len-ae.size))

    def getMFCCNorm(mfcc):
        norm = speechpy.processing.cmvn(mfcc.T, variance_normalization=True)
        return norm.T

    def DTW(ref_mfcc, comp_mfcc):
        D, wp = librosa.sequence.dtw(X=ref_mfcc, Y=comp_mfcc, metric='euclidean')
        wp_s = np.asarray(wp) * hop_length / sr
        return D, wp, wp_s
    
    def plotDTW(D, wp_s):
        fig = plt.figure(figsize=(10, 7))
        ax = fig.add_subplot(111)
        librosa.display.specshow(D, x_axis='time', y_axis='time',
                                cmap='gray_r', hop_length=hop_length)
        ax.plot(wp_s[:, 1], wp_s[:, 0], marker='o', color='r')
        plt.rcParams['axes.grid']=False
        plt.colorbar()
        plt.title('Warping Path on Acc. Cost Matrix $D$')
        plt.show()

    def getDiff(ref_mfcc, comp_mfcc, wp, sr, hop_length, n_fft, ref_ae, comp_ae):
        length = max(ref_mfcc.shape[1], comp_mfcc.shape[1])
        x = np.linspace(0, length-1, length)
        x_time = librosa.frames_to_time(x, sr=sr, hop_length=hop_length, n_fft = n_fft)
        re_wp = wp[::-1]
        y = []

        ref_mm = minmax_scale(ref_mfcc, axis=0, copy=True)
        comp_mm = minmax_scale(comp_mfcc, axis=0, copy=True)
        
        ref_mm_mean = []
        comp_mm_mean = []
        
        for i in range(ref_mfcc.shape[1]):
            ref_mm_mean.append(ref_mm[:,i].mean())
        for i in range(comp_mfcc.shape[1]):
            comp_mm_mean.append(comp_mm[:,i].mean())
        
        ref_mm_mean = np.array(ref_mm_mean)
        comp_mm_mean = np.array(comp_mm_mean)

        passed = 0
        for ref_fr, comp_fr in re_wp:
            if len(y) >= length:
                passed += 1
                continue
            val = abs(ref_mm_mean[ref_fr] - comp_mm_mean[comp_fr])/ (ref_mm_mean[ref_fr] if ref_mm_mean[ref_fr] > 0 else 1)
            y_val = min(1, val)
            y.append(y_val)
        print("passed y: ", passed)

        if comp_ae.size < x.size:
            comp_ae = pad(comp_ae, x.size)
        if ref_ae.size < x.size:
            ref_ae = pad(ref_ae, x.size)

        for idx in range(len(y)):
            if (ref_ae[idx] == comp_ae[idx]) and (ref_ae[max((idx-1),0)] == comp_ae[max(idx-1,0)]) and (ref_ae[max((idx-2),0)] == comp_ae[max(idx-2,0)]):
                y[idx] = 0

        return x_time, y, re_wp, ref_ae, comp_ae, ref_mm_mean, comp_mm_mean
    
    def getPDiff(ref_mfcc_norm, comp_mfcc_norm, re_wp):
        p_diff = []
        for ref_fr, comp_fr in re_wp:
            a,b = scipy.stats.pearsonr(ref_mfcc_norm[:,ref_fr], comp_mfcc_norm[:,comp_fr])
            a = abs(a)
            p_diff.append(round(a,4))
        p_mean = np.mean(p_diff)
        return p_mean
    
    def timeAlign(ref_trim, comp_trim, sr):
        flag = 1
        pad_num=0
        # where comparison signal is longer than the reference
        if comp_trim.shape[0] > ref_trim.shape[0]:
            c = scipy.signal.correlate(comp_trim, ref_trim, mode="valid", method="fft")
            peak = np.argmax(c)
            if int(round(peak,0)) + ref_trim.shape[0] < comp_trim.shape[0]:
                pad_num = int(round(peak,0))
            else:
                pad_num = comp_trim.shape[0] - ref_trim.shape[0]
            ref_trim = np.pad(ref_trim, (pad_num,0), 'linear_ramp', end_values=(0,0))
            flag = 0
        
        elif comp_trim.shape[0] < ref_trim.shape[0]:
            c = scipy.signal.correlate(ref_trim, comp_trim, mode="valid", method="fft")
            peak = np.argmax(c)
            if int(round(peak,0)) + comp_trim.shape[0] < ref_trim.shape[0]:
                pad_num = int(round(peak,0))
            else:
                pad_num = ref_trim.shape[0] - comp_trim.shape[0]
            comp_trim = np.pad(comp_trim, (pad_num,0), 'linear_ramp', end_values=(0,0))
            flag = 1
        time_lag = pad_num/sr
        return ref_trim, comp_trim, time_lag, flag
    
    def plotAlign(ref, comp, x, y, ref_ae, comp_ae, time_lag, flag):
        fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, sharex=True, figsize=(13, 15))
        
        # Plot x_2
        librosa.display.waveshow(comp, sr=sr, ax=ax2)
        ax2.set(title='comparison waveplot')
        ax2.grid(True, axis='both')

        # Plot x_1
        librosa.display.waveshow(ref, sr=sr, ax=ax1)
        ax1.set(title='reference waveplot')
        ax1.label_outer()
        ax1.grid(True, axis='both')
        
        n_arrows = 20
        if flag == 0:
            for tp1, tp2 in wp_s[::len(wp_s)//n_arrows]:
                tp1 += time_lag
                con = ConnectionPatch(xyA=(tp1, 0), xyB=(tp2, 0),
                                    axesA=ax1, axesB=ax2,
                                    coordsA='data', coordsB='data',
                                    color='r', linestyle='--',
                                    alpha=0.5)
                ax2.add_artist(con)
        
        else:
            for tp1, tp2 in wp_s[::len(wp_s)//n_arrows]:
                tp2 += time_lag
                con = ConnectionPatch(xyA=(tp1, 0), xyB=(tp2, 0),
                                    axesA=ax1, axesB=ax2,
                                    coordsA='data', coordsB='data',
                                    color='r', linestyle='--',
                                    alpha=0.5)
                ax2.add_artist(con)

        t_x = x[::3]
        t_y = []
        t_y.append(y[0])
        for i in range(0,x.size-3,3):
            t_y.append(mean(y[i:i+6]))
            
        ax3.set(title='ref v comparison')
        ax3.grid(True, axis='both')
        ax3.plot(x,ref_ae, label="ref", color = 'darkblue')
        ax3.plot(x,comp_ae, label="comp", color = 'skyblue')
        ax3.plot(t_x,t_y, label="MFCC_diff", color = 'r', linestyle='--')
        ax3.set_xlabel("Time")
        ax3.set_ylim(ymin=0, ymax=1)
        ax3.legend()
        plt.show()
    
    def getPhoneme(pointers):
        pointers = pointers
        pointers.append(librosa.get_duration(y=y_ref))

        pointers_comp = []
        for x in pointers:
            data_interval = wp_s[np.where(wp_s[:,0] <= x)]
            max_interval = data_interval.max(axis=0)[1]
            pointers_comp.append(max_interval)
        
        return pointers, pointers_comp

    def plotPhoneme(pointers, pointers_comp):
        fig, ax = plt.subplots(nrows=2, sharex=True, figsize=(10,8))
        librosa.display.waveshow(y_ref, sr=sr, ax=ax[0])
        ax[0].vlines(pointers, -max(abs(y_ref)), max(abs(y_ref)), color="r", linestyles='--')
        ax[0].set(title='Reference Waveplot')
        ax[0].label_outer()
        librosa.display.waveshow(y_comp_trim, sr=sr, ax=ax[1])
        ax[1].vlines(pointers_comp, -max(abs(y_comp_trim)), max(abs(y_comp_trim)), color="r", linestyles='--')
        ax[1].set(title='Comparison Waveplot')
        plt.show()

    def getPhonemeTime(pointers, pointers_comp, ref_mfcc, comp_mfcc, sr, hop_length):
        fr_ref = librosa.time_to_frames(pointers, sr=sr, hop_length=hop_length)
        fr_ref[-1] = ref_mfcc.shape[1]
        fr_comp = librosa.time_to_frames(pointers_comp, sr=sr, hop_length=hop_length)
        fr_comp[-1] = comp_mfcc.shape[1]

        return fr_ref, fr_comp

    def getDiffPhoneme(ref_mm_mean, comp_mm_mean, word, fr_ref, fr_comp):
        for i in range(len(word)):
            a = ref_mm_mean[fr_ref[i]:fr_ref[i+1]+1].mean()
            b = comp_mm_mean[fr_comp[i]:fr_comp[i+1]+1].mean()
            print(word[i], ": ", "Ref 평균값: ", round(a,5), ", Comp 평균값: ", round(b,5), ", Difference (%): ", round((abs(a-b)/(a if a > 0 else 1)*100),2), "%")

            fr_diff_ref = fr_ref[i+1] - fr_ref[i]
            fr_diff_comp = fr_comp[i+1] - fr_comp[i]

            if fr_diff_ref - fr_diff_comp >= 6:
                print(f'"{word[i]}" 음절은 조금 천천히 얘기해주세요')
            
            elif fr_diff_ref - fr_diff_comp <= -6:
                print(f'"{word[i]}" 음절은 조금 빠르게 얘기해주세요')

    plt.style.use('seaborn-muted')

    word = list(word)

    y_ref, sr = librosa.load(path+ref, mono=True, sr=sr)
    y_comp, _ = librosa.load(path+target, mono=True, sr=sr)

    hop_length = 128
    n_fft = 512

    y_ref_trim = y_ref
    y_comp_trim = getTrim(y_comp)

    ref_mfcc = getMFCC(y_ref_trim, sr=sr, hop_length=hop_length, n_fft=n_fft, n_mfcc=13)
    comp_mfcc = getMFCC(y_comp_trim, sr=sr, hop_length=hop_length, n_fft=n_fft, n_mfcc=13)

    ref_mfcc_norm = getMFCCNorm(ref_mfcc)
    comp_mfcc_norm = getMFCCNorm(comp_mfcc)

    D, wp, wp_s = DTW(ref_mfcc_norm, comp_mfcc_norm)

    plotDTW(D, wp_s)

    y_ref_trim, y_comp_trim, time_lag, flag = timeAlign(y_ref_trim, y_comp_trim, sr)
    ref_ae = amplitude_envelope(y_ref_trim, n_fft, hop_length)
    comp_ae = amplitude_envelope(y_comp_trim, n_fft, hop_length)
    x_time, y, re_wp, ref_ae, comp_ae, ref_mm_mean, comp_mm_mean = getDiff(ref_mfcc, comp_mfcc, wp, sr, hop_length, n_fft, ref_ae, comp_ae)
    plotAlign(y_ref_trim, y_comp_trim, x_time, y, ref_ae, comp_ae, time_lag, flag)
    p_mean = getPDiff(ref_mfcc_norm, comp_mfcc_norm, re_wp)
    print("Pearson Similarity: ", round(p_mean*100,2), "%")

    pointers, pointers_comp = getPhoneme(phoneme)
    plotPhoneme(pointers, pointers_comp)
    fr_ref, fr_comp = getPhonemeTime(pointers, pointers_comp, ref_mfcc, comp_mfcc, sr, hop_length)
    getDiffPhoneme(ref_mm_mean, comp_mm_mean, word, fr_ref, fr_comp)

    return
```
